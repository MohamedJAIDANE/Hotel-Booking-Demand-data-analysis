# Hotel Booking Demand
This dataset consists of booking data from a city hotel and a resort hotel. It includes many details about the bookings, including room specifications, the length of stay, the time between the booking and the stay, whether the booking was canceled, and how the booking was made. The data was gathered between July 2015 and August 2017.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px


booking=pd.read_csv("hotel_bookings_clean.csv",index_col=False)
```

```python
booking['id']=booking.index
booking.set_index('id')
```

```python
by_agent=booking["booked_by_agent"] == 1
booking["family_size"]=booking["adults"]+booking["children"]+booking["babies"]
```

```python
booking.info()
```

```python
df_results = booking[(booking.booked_by_agent == True) & (booking.is_canceled == 1)&(booking.family_size)]
df_results.describe()
```

```python
cancellation_family_size = df_results.groupby(['family_size'])['id'].count()
cancellation_family_size.index=["1p","2p","3p","4p","5p","12p","20p"]
```

```python
cancellation_family_size.plot(kind='bar')
plt.xlabel('FamilySize')
plt.ylabel('Total')
plt.title('Family size Cancelation')
plt.show()
```

```python
reservations_by_month = booking.groupby('arrival_date_month')['id'].count()
reservations_by_month
reservations_by_month.index=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'] 
```

```python
plt.rcParams['figure.figsize']=(20,6)
reservations_by_month.plot(kind='bar')
plt.xlabel("months")
plt.ylabel('Nbr de réservation')
plt.title("reservation by month")
```

```python
canceled_df=booking[(booking.is_canceled == 1)]
canceled_df
```

```python
cancellation_rates_month=canceled_df.groupby(['arrival_date_month'])['id'].count()
cancellation_rates_month.index=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'] 
```

```python
cancellation_rates_month.plot(kind='bar')
plt.xlabel('Months')
plt.ylabel('Total')
plt.title('cancellation rates by month')
```
