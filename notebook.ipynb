The data was gathered between July 2015 and August 2017.

## Question / Objectif
<strong>Quelles tailles de famille sont associ√©es au taux d'annulation le plus √©lev√© ?<strong>

on va changer l'index de booking

```python
booking['id']=booking.index
booking.set_index('id')
```

### Maintenant nous devons calculer la taille de la famille  
c'est la somme des enfants, des b√©b√©s et des adultes pour obtenir le r√©sultat et ajouter une colonne appel√©e family_size.

```python
booking["family_size"]=booking["adults"]+booking["children"]+booking["babies"]
```

```python
booking['family_size'].describe()
```

# Remarque
### Il ya des outliers
pour d√©tecter les outLiner on va utiliser boxplot

#### Visualisation

```python
test=pd.DataFrame(booking['family_size'])
test.boxplot()
plt.show()
```

### on utiliser les valeur inf√©rieur ou √©gale √† 10 p

Les autre valeur se sont des cas exeptionel ou il ya par exemple un evenement 

donc on va choisir les valeur qui ont inf√©rieur ou √©gale a 10 pour eviter les outliner

```python
df_results = booking[(booking.is_canceled == 1)&(booking.family_size<=10)]
df_results.describe()
```

## Quelles tailles de famille sont associ√©es au taux d'annulation le plus √©lev√© ?
* Utilisons groupby, qui permet de produire des tables de synth√®ses par cat√©gories:

```python
cancellation_family_size = pd.DataFrame(df_results.groupby(['family_size'])['id'].count())
cancellation_family_size.index
```

```python
cancellation_family_size['%']=cancellation_family_size['id']/cancellation_family_size['id'].sum()*100
cancellation_family_size
```

Changer l'index

```python
cancellation_family_size.index=['1P','2P','3P','4P','5P','6P','10P']
```

```python
cancellation_family_size.plot(kind='bar')
plt.xlabel('FamilySize')
plt.ylabel('Total')
plt.title('Family size Cancelation')
plt.show()
```

#### Conclution
* Observation :

  * 73,71% des Annulation sont annul√©e par deux personnes
  

* Interpr√©tation : La famille de taille 2 eu plus de chances d'annulation que les autres

Pour aller plus loin, on pourrait regarder les taux d'annulation √† diff√©rentes p√©riodes de l'ann√©e.

# les taux d'annulation √† diff√©rentes p√©riodes de l'ann√©e.

Nous observerons les taux de r√©servation par rapport au mois

```python
reservations_by_month = pd.DataFrame(booking.groupby(['arrival_date_month'])['id'].count())
reservations_by_month.index=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'] 
reservations_by_month.plot(kind='bar')
```

```python
plt.rcParams['figure.figsize']=(20,6)
reservations_by_month.plot(kind='bar')
plt.xlabel("months")
plt.ylabel('Nbr de r√©servation')
plt.title("reservation by month")
```

Nous observerons les taux des annulations par rapport au mois

```python
canceled_df=booking[(booking.is_canceled == 1)]
canceled_df
```

```python
cancellation_rates_month=canceled_df.groupby('arrival_date_month')['id'].count()
cancellation_rates_month.index=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'] 
```

```python
cancellation_rates_month.plot(kind='bar')
plt.xlabel('Months')
plt.ylabel('Total')
plt.title('cancellation rates by month')
```

#### Conclution
* Observation :

  * On peut remarquer qu'au moins en ao√ªt les r√©servations augmentent 
  
* Interpr√©tation : en ao√ªt, le taux d'annulation augmente, puisque le nombre de r√©servations augmente √©galement.

Pour aller plus loin, on pourrait regarder l'influence du d√©lai d'ex√©cution sur les annulation


## Les r√©servations dont le d√©lai d'ex√©cution est plus long sont-elles plus susceptibles d'√™tre annul√©es ?

```python
cancellation_rates_leadtime=canceled_df.groupby(['lead_time'])['id'].count()
cancellation_rates_leadtime.plot(kind='line')
```

#### Conclusion
